## Conversation between chatbots

```python
gemini = "gemini-2.5-flash"
llama = "llama2:7b"


gemini_system = "You are a chatbot who wants to start a freelance business."

llama_system = "You are a very deep thinking chatbot that gives answer and roadmaps to the asked questions."

# gemini_message = ["Hi there"]
# llama_message = ["Hi"]
```

```python
def call_gemini(gemini_hist, llama_hist):
    messages = [{"role": "system", "content": gemini_system}]
    for gemini_msg, llama_msg in zip(gemini_hist, llama_hist):
        messages.append({"role": "assistant", "content": gemini_msg})
        messages.append({"role": "user", "content": llama_msg})
    completion = gemini_via_openai_client.chat.completions.create(
        model=gemini,
        messages=messages
    )
    return completion.choices[0].message.content
```

```python
# You need to pass two arguments: gemini_hist and llama_hist
# For example, to start a conversation:
gemini_hist = ["Hi there"]
llama_hist = ["Hello!"]

call_gemini(gemini_hist, llama_hist)
```

```python
def call_llama(gemini_hist, llama_hist):
    messages = []
    for gemini_msg, llama_msg in zip(gemini_hist, llama_hist):
        messages.append({"role": "user", "content": gemini_msg})
        messages.append({"role": "assistant", "content": llama_msg})
    messages.append({"role": "user", "content": gemini_hist[-1]})
    payload = {
        "model": llama,
        "messages": messages,
        "stream": False,
        "system": llama_system
    }
    response = requests.post(ollama_api, headers=headers, json=payload)
    result = response.json()
    return result["message"]["content"]
```

```python
gemini_hist = ["What is your favorite programming language?"]
llama_hist = ["I love Python! It's very versatile and great for many applications."]
call_llama(gemini_hist, llama_hist)
```

```python
gemini_history = ["Hi Llama, i want to start a freelance business. Like i want to create a digital product and sell it online. like: an ai created animation video. \
                    Give me a roadmap how can i start to create videos and give one sample prompt to create a video."]
llama_history = [] # Llama hasn't spoken yet

print("--- Starting Conversation ---")
print(f"Gemini (Turn 0):\n{gemini_history[0]}\n")

for i in range(5): # Let's do 5 turns for the example
    # 1. Llama answers Gemini's last question
    # We pass the full history to the function
    llama_nxt = call_llama(gemini_history, llama_history)
    print(f"Llama (Turn {i}):\n{llama_nxt}\n")
    llama_history.append(llama_nxt)
    
    # 2. Gemini asks a follow-up question based on Llama's answer
    # We pass the full history to the function
    gemini_nxt = call_gemini(gemini_history, llama_history)
    print(f"Gemini (Turn {i+1}):\n{gemini_nxt}\n")
    gemini_history.append(gemini_nxt)

print("--- Conversation Ended ---")
```
